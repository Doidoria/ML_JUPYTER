```python
#--------------------------
# 상관관계 확인(수치형)
#--------------------------
import pandas as pd

data = {
    "키":[150,160,170,175,180],
    "몸무게":[40,50,60,70,80]
}

df=pd.DataFrame(data)
df
```

```python
# 상관계수 확인(기본 : 피어슨 상관계수 -1 ~ 1 )
# +1 에 가까울수록 양의 상관관계
# -1 에 가까울수록 음의 상관관계
# 0 에 가까울수록 상관이 거이없음

correlation=df.corr()
print(correlation)
```

```python
# 피어슨 상관계수(기본) - 선형관계가 얼마나 강한지 확인 / 곡선관계 확인은 힘들다
print(df.corr())

# 스피어맨 상관계수 - 순위(등수)를 기준으로 비교하는 방법 / 곡선관계도 확인 가능
print(df.corr(method='spearman'))

#켄달타우 상관계수 - 순위(등수)를 기준으로 비교하는 방법(순서쌍을 만들어서 비교) / 곡선관계도 확인 가능 (좀 더 정밀하다)
print(df.corr(method='kendall'))
```
