```python
# ---------------------
# 데이터 전처리
# ---------------------
```

```python
#---------------------
# 인코딩[범주형 -> 수치형]
#---------------------
# 라벨(정답,y값) 분리
y_train = train.pop("Credit_Score")
```

```python
y_train.value_counts()
```

```python
# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

cols=train.columns[train.dtypes==object]

cols
for col in cols:
    le = LabelEncoder()
    le.fit(pd.concat([train[col],test[col]],axis=0)) # 행 합치기
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
```

```python
train
```

```python
test
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val=train_test_split(
    train,
    y_train,
    test_size=0.2,
    random_state=0
)
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape # x - 학습, y - 검증
```

```python
# ---------------------
# 머신러닝 학습 및 평가
# ---------------------
```

```python
# 랜덤 포레스트 학습
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0) # 난수값고정 in -> out 고정값
rf.fit(X_train, y_train) # 모델학습
```

```python
# LIGHTGBM 학습
import lightgbm as lgb
lgbmc = lgb.LGBMClassifier(random_state=0, verbose=-1)
lgbmc.fit(X_train, y_train)
pred = lgbmc.predict(X_val) # 예측 y 값
pred_proba = lgbmc.predict_proba(X_val) # 예측 확률값
```

```python
# 평가
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

# roc_auc = roc_auc_score(y_val, pred_proba[:,1]) # pred 예측 확률값 [y_val - 오답예측확률, pred_proba - 정답예측확률]
accuracy = accuracy_score(y_val, pred) # pred 예측 y 값
f1 = f1_score(y_val, pred, average='macro') # pred 예측 y 값 (클래스별(1,2,3정답값) 성능을 동등하게 반영할때 사용)

# print('roc_auc : ', roc_auc)
print('accuracy : ', accuracy)
print('f1 : ', f1)
```
